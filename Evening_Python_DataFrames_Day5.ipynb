Data Filtering

```python
# Display startups belonging to NewYork
df[df['STATE']=='New York']
```

```python
# Display records of startups where Profit is greater than or equal to 50000
df[df['PROFIT']>=50000]
```

```python
# display startups from Florida where MKT expense is greater than or equal to 100000
df[(df['STATE']=='Florida')& (df['MKT']>=100000)]
```

```python
# display startups where profit is in between 100000 to 150000 &
df[(df['PROFIT']>=100000) & (df['PROFIT']<=150000)]
```

```python
# display startups where profit is in between 100000 to 150000 and belongs to California
df[((df['PROFIT']>=100000) & (df['PROFIT']<=150000)) & (df['STATE']=='California')]
```

```python
# Sort the records
# dataframe.sort_values(by='colname') # sorts in ascending order
# dataframe.sort_values(by='colname',ascending=False) #sorts in descending
```

```python
# arrange records based on PROFIT
df.sort_values(by='PROFIT')
```

```python
# display top 5 startups based on Profit
df.sort_values(by='PROFIT',ascending=False).head(5)
```
